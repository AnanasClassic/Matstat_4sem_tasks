```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
n = 1000
theta = stats.uniform.rvs(0, 20)
theta
```

```python
X = np.array(stats.uniform.rvs(0, theta, n))
plt.hist(X)
plt.show()
```

### оптимальная оценка: $\hat{\theta_1} = \frac{n + 1}{n} X_{(n)}$
### оценка по методу моментов: $\hat{\theta_2} = 2\overline{X}$

```python
theta_1 = [np.max(X) * (n + 1) / n]
theta_1[0]
```

```python
theta_2 = [np.mean(X) * 2]
theta_2[0]
```

```python
for i in range (100):
    X = np.array(stats.uniform.rvs(0, theta, n))
    theta_1.append(np.max(X) * (n + 1) / n)
    theta_2.append(np.mean(X) * 2)
```

```python
plt.hist(theta_1, bins=50)
plt.axvline(x = theta, color='red')
plt.show()
```

```python
plt.hist(theta_2, bins=50)
plt.axvline(x = theta, color='red')
plt.show()
```

```python
print(np.var(theta_1))
print(np.var(theta_2))
```

```python
print(abs(np.mean(theta_1 - theta)))
print(abs(np.mean(theta_2 - theta)))
```

```python
n = 1000
for i in range(2):
    theta = stats.uniform.rvs(0, 20)
    theta_1 = []
    theta_2 = []
    for k in range (1001):
        X = np.array(stats.uniform.rvs(0, theta, n))
        theta_1.append(np.max(X) * (n + 1) / n)
        theta_2.append(np.mean(X) * 2)
    plt.hist((theta_1 - theta) * np.sqrt(n), bins=50)
    # plt.axvline(x = theta, color='red')
    plt.show()
    plt.hist((theta_2 - theta) * np.sqrt(n), bins=50)
    # plt.axvline(x = theta, color='red')
    plt.show()
    print(np.var(theta_1), abs(np.mean(theta_1 - theta)))
    print(np.var(theta_2), abs(np.mean(theta_2 - theta)))
```


### $\mathbb{E} X_{(n)}^2 = \int_{0}^{\theta} nx^2 \frac{x^{n - 1}}{\theta^n}dx = \frac{n \theta^2}{n + 2}$
### $\mathbb{D} \frac {n + 1}{n} X_{(n)} = \frac {n + 1}{n} \left( \frac{\theta^2n}{(n + 2)} - \frac{\theta^2 n^2}{(n + 1)^2}\right) = \frac{n + 1}{n} \theta^2 \frac{n}{(n + 1)^2(n + 2)} = \frac{\theta^2}{(n + 1)(n + 2)}$

### $\mathbb{D}(2 \overline{X}) = \frac{4}{n} \mathbb{D} X_1 =  \frac{\theta^2}{3n}$

### сравнивая с $\frac{1}{ni(\theta)}$, получаем, что оптимальная оценка сходится с лучшей асимптотикой, а оценка по методу моментов с той же
